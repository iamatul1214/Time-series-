```python
## A quick function to check how to calculate the time taken by a function to execute
def check_time():
    start_time=datetime.now()
    sum=1
    print(f"Function started on {start_time}")
    for i in range(1,15,1):
          sum=sum*i
    print(f"the sum ={sum}")
              
          
    end_time=datetime.now()
    print(f"Function ended on {end_time}")     
    print(f"The total time taken to execute the function is {end_time-start_time}")
          
          
    
```

```python
check_time()
```

## Time resampling

```python
df_testla.head()
```

```python
df_tesla=df_testla.set_index('Date')
df_tesla.head()
```

```python
df_tesla.resample(rule='A').min() ## This shows the min values of all the features for each year. rule= A will resample data on yearly basis
```

```python
df_tesla.resample(rule='B').min()   ##Rule=B stands for Business days
```

```python
df_tesla.resample(rule='C').min()     ## Rule C stands for custom business day
```

```python
df_tesla.resample(rule='W').min() ## Rule W stands for week
```

```python
df_tesla.resample(rule='M').min()     ## Rule M stands for calender MOnth end
```

```python
### Now let's try to plot some information regarding Maximum open price on yearly basis and then on monthly basis
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
df_tesla.resample(rule='M').max()['Open'].plot()
```

```python
## We can have rule=QS for quaterly 
df_tesla.resample(rule='QS').max()['Open'].plot()
```

```python
## Let's use subplots to plot these observations
figure, ax=plt.subplots(1,3, figsize=(20,5))
figure.suptitle("A graph between Max open vs the number")
ax[0].plot(df_tesla.resample(rule='M').max()['Open'])
ax[0].set_title('Monthly Max Open')
ax[1].plot(df_tesla.resample(rule='Y').max()['Open'])
ax[1].set_title('Yearly Max Open')
ax[2].plot(df_tesla.resample(rule='QS').max()['Open'])
ax[2].set_title('Quaterly Max Open')
```

```python
figure, ax=plt.subplots(1,3, figsize=(20,5))
figure.suptitle("A graph between Max High vs the number")
ax[0].plot(df_tesla.resample(rule='M').max()['High'])
ax[0].set_title('Monthly Max High')
ax[1].plot(df_tesla.resample(rule='Y').max()['High'])
ax[1].set_title('Yearly Max High')
ax[2].plot(df_tesla.resample(rule='QS').max()['High'])
ax[2].set_title('Quaterly Max High')
```

```python
figure, ax=plt.subplots(1,3, figsize=(20,5))
figure.suptitle("A graph between Max Low vs the number")
ax[0].plot(df_tesla.resample(rule='M').max()['Low'])
ax[0].set_title('Monthly Max Low')
ax[1].plot(df_tesla.resample(rule='Y').max()['Low'])
ax[1].set_title('Yearly Max Low')
ax[2].plot(df_tesla.resample(rule='QS').max()['Low'])
ax[2].set_title('Quaterly Max Low')
```

```python
## Now Let's see the bar plots 
df_tesla.resample(rule='A').max()['Open'].plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar', figsize=(10,6))
```

```python
df_tesla['Open'].resample(rule='M').mean().plot(kind='bar', figsize=(20,6))
```

```python
df_tesla['Open'].resample(rule='QS').mean().plot(kind='bar', figsize=(20,6))
```

### Rolling and smoothing

```python
df_tesla['High'].rolling(11).max().head(20)
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head(40)
```

```python
## Let's try to plot open vs the new column
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,5))
```

### On the above plot we can see that orange line starts after some point of blue line because it has some Nan values due to ROLLING. This yellow line is termed as smoothing, because we have smoothened the too much zig zag of blue line with the help of yellow line by taking the mean.
